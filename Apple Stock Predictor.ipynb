```python
import yfinance as yf
```

```python
ap=yf.Ticker("AAPL")
```

```python
ap=ap.history(period='max')
```

```python
ap
```

```python
ap.index
ap.plot(y='Close',use_index='True')
```

```python
del ap['Dividends']
del ap['Stock Splits']
```

```python
ap['Tomorrow']=ap['Close'].shift(-1)
ap
```

```python
ap['Increase']=(ap['Tomorrow']>ap['Close']).astype(int)
ap
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
train=ap.iloc[:-100]
test=ap.iloc[-100:]
predictors = ['Close', 'Volume', 'Open', 'High', 'Low']
model.fit(train[predictors], train['Increase'])
```

```python
from sklearn.metrics import precision_score
pred = model.predict(test[predictors])
```

```python
import pandas as pd
pred = pd.Series(pred, index=test.index)
precision_score(test['Increase'], pred)
```

```python
combined = pd.concat([test['Increase'], pred], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Increase'])
    pred= model.predict(test[predictors])
    pred= pd.Series(pred, index=test.index, name="Predictions")
    combined = pd.concat([test['Increase'], pred], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions=predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(ap, model, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Increase'], predictions['Predictions'])
```

```python
predictions['Increase'].value_counts()/predictions.shape[0]
```

```python
 horizons=[2,5,60,250,1000]
```

```python
new_predictors = []
for horizon in horizons:
    rolling = ap.rolling(horizon).mean()
    ratio_column = f"Close_Ratio{horizon}"
    ap[ratio_column]=ap['Close']/rolling['Close']
    trend_column=f"Trend_{horizon}"
    ap[trend_column]=ap.shift(1).rolling(horizon).sum()['Increase']
    new_predictors += [ratio_column, trend_column]
```

```python
ap = ap.dropna()
ap
```

```python
model = RandomForestClassifier(n_estimators=1000, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Increase'])
    pred= model.predict_proba(test[predictors])[:,1]
    pred[pred>=.6]=1
    pred[pred<.6]=0
    pred= pd.Series(pred, index=test.index, name="Predictions")
    combined = pd.concat([test['Increase'], pred], axis=1)
    return combined
```

```python
predictions = backtest(ap, model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Increase'], predictions['Predictions'])
```
